# Data Cleaning in Pandas 

```python
# importing the pandas library
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\MR.OKECH\Desktop\Joshua\Data Career\Data Analyst Bootcamp\Bootcamp Files\Pandas\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
# df["Last_Name"] = df["Last_Name"].str.lstrip("...")
# df["Last_Name"] = df["Last_Name"].str.lstrip("/")
# df["Last_Name"] = df["Last_Name"].str.rstrip("_")

df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
df
```

```python
# df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex = True)

# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df["Phone_Number"] = df["Phone_Number"].str.strip().str.replace(r'nan--\s*', '', regex=True)
df["Phone_Number"] = df["Phone_Number"].str.strip().str.replace(r'Na--\s*', '', regex=True)

df
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', n=2, expand=True)
df
```

```python
# df['Paying Customer'] = df['Paying Customer'].str.replace('N', 'No')
# df['Paying Customer'] = df['Paying Customer'].str.replace('Y', 'Yes')

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('N', 'No')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Y', 'Yes')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No/a', 'N/a')

df
```

```python
df = df.replace('N/a', '')
df = df.replace('NaN', '')
df = df.replace('No/a', 'No')

df = df.map(lambda x: pd.NA if x is None else x)
df = df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Yes':
        df.drop(x, inplace = True)

df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('', 'No')
df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace = True)

df
```

```python
df = df.reset_index(drop = True)
df
```
